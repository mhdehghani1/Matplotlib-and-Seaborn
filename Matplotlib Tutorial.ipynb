## Question 01




#### Question 01.a

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

%matplotlib inline

y=[30,24,8,38]
x=['A','B','AB','O']

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=8)





plt.title('KHOON')
plt.xlabel('گروه خونی')
plt.ylabel('تعداد')
plt.show()
```

#### Question 01.b

```python
# plot
fig, ax = plt.subplots()

ax.pie(y, radius=4, center=(4, 4),labels=x,labeldistance=.7, wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)
plt.title('KHOON')
plt.xlabel('گروه خونی')
plt.ylabel('تعداد')


plt.show()
```

#### Question 01.c

```python
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=8)



plt.style.use('dark_background')


plt.title('KHOON')
plt.xlabel('گروه خونی')
plt.ylabel('تعداد')

plt.show()
```

## Question 02




```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

# plot
fig=plt.figure(dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.scatter(x, y)

plt.title('CITY')
plt.xlabel('cars')
plt.ylabel('person')

plt.show()
```
