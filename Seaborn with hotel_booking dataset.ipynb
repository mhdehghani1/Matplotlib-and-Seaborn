## Question 04




```python
col1=(df['arrival_date_year'])
col2=(df['arrival_date_month'])
col1=col1.values
col2=col2.values
dfNew=pd.DataFrame()
dfNew['col1']=col1
dfNew['col2']=col2
dfNew['tt']=dfNew.groupby('col2')['col1'].transform(lambda x:x.count())
dfNew
tt=dfNew.groupby('col2')['col1'].count()
fp=dfNew.pivot_table(index='col1', columns='col2',values='tt')
fp
```

```python
sns.heatmap(fp, cmap='Wistia', linecolor='black', linewidths=2)
```

**Your Answer:**در سال 2016 تعدا مسافر بیشتری بوده است

## Question 05




```python
col1=(df['is_canceled']==0).astype(int)
col2=(df['is_canceled']==1).astype(int)
col1=col1.values
col2=col2.values
dfkde=pd.DataFrame()
dfkde['col1']=col1
dfkde['col2']=col2

sns.kdeplot(data=dfkde, x='col1', label='Not_Cancel' )
sns.kdeplot(data=dfkde,x='col2', label= 'Is_Cancel')

plt.legend(loc=(1.1, 0.5))
```

**Your Answer:**کنسلی ها و رزورها تقریبا برابر است و از هر دو مشتری یکی کنسل میکند
